- Here are lots of features here. In total, our training data has 59 variables, including id and target. In some of them we already see a number of NAs. 

- The data description mentions that the names of the features indicate whether they are binary (bin) or categorical (cat) variables. Everything else is continuous or ordinal.

- We have already been told by Adriano Moala that the names of the variables indicate certain properties: *Ind" is related to individual or driver, “reg” is related to region, “car” is related to car itself and “calc” is an calculated feature.’ Here we will refer to these properties as groups.

These names refer to the origin of variable, but there is no reason to concern about it. "Ind" is related to individual or driver, "reg" is related to region, "car" is related to car itself and "calc" is an calculated feature. The features and rows are not time dependent, so "ps_ind_01" and "ps_ind_02" are just labels to hide real names. Unfortunately I can't share the real meanings of variables.

The 1st place solution deletes all the calc features from the remaining features. 

Get all the feature name list 

````python
features = train.columns
````

````python
len(features)
````

````python
features[3]
````

````python
def calculatedFetures(f):
    calc_features = []
    for feature in f:
        if "calc" in feature:
            calc_features.append(feature)
    return calc_features

calc_features = calculatedFetures(features)
````

````python
def getCoreFeatures(first, second):
    second = set(second)   
    return [item for item in first if item not in second]

core_features = getCoreFeatures(features, calc_features)    
````

````python
core_features
````

````python
len(core_features),len(calc_features) #sanity check that we are missing any other feature
````

So we will be working mainly with core_features as mentioned in the solutioning

````python
core_data = train[core_features] ### This is where i get rid of *calc features 
````

````python
DataFrameSummary(core_data).summary()
````

Turning categorical variables and binary variables into one hot encoding 

Get all cat variable

````python
def catFetures(f):
    cat_features = []
    for feature in f:
        if "cat" in feature:
            cat_features.append(feature)
    return cat_features
````

````python
cat_features = catFetures(core_data.columns)
````

````python
cat_features
````

````python
len(cat_features)
````

````python
def binFetures(f):
    bin_features = []
    for feature in f:
        if "bin" in feature:
            bin_features.append(feature)
    return bin_features
````

````python
bin_features = binFetures(core_data.columns)
````

````python
bin_features
````

````python
len(bin_features)
````

````python
cat_columns = cat_features + bin_features
````

````python
print(cat_columns)
````

````python
len(cat_columns)
````

````python
def one_hot_encode(core_data,cat_columns):    
    for cat_column in cat_columns: 
        print("Resolving columns", cat_column)
        core_data_with_onehot = pd.get_dummies(core_data[cat_column])
        core_data = pd.concat([core_data, core_data_with_onehot], axis=1)
        del core_data[cat_column]    
    return core_data

core_data = one_hot_encode(core_data, cat_columns)
````
